```python
# Importanto bibliotecas
import numpy as np
import pandas as pd
```

Portanto em um primeiro momento devemos carregar o _dataset_ conforme procedimento descrito na célula abaixo. Após o upload do _dataset_ aplicou-se o método `.info()` apresenta um breve resumo dos valores contidos nesse _Data Frame_ denominado **base_carros**. São extraídas informações como tipagem da variável, _id_ da variável e quantidade de valores preenchidos ( _non-null_ ) por exemplo.

```python
caminho = 'C:/Users/Usuario/Documents/PythonFiles/vehicles.csv'
base_carros = pd.read_csv(caminho)
base_carros.info()
```

Para um visualização parcial dos dados podemos usar o comando `.head()` combinado ao método `pd.set_option("display.max_columns", digitar aqui valor de colunas desejadas)`. O método `.head()` por _default_ apresentará apenas os 5 primeiros registros. Para alterar basta escrever a quantidade de registros desejados dentro do `()`. 

Nesse caso desejamos ver as 5 primieras 25 colunas do _dataset_ no console _notebook jupyter_.

```python
pd.set_option("display.max_columns", 25)
base_carros.head()
```

Após essa visualização preliminar dos dados e com auxílio da biblioteca pandas poderemos eliminar as colunas ou linhas de registros indesejados que por exemplo não teriam impacto numa análise de predição. No caso do exemplo dos carros vamos eliminar as colunas `'url'`, `'region'`, `'region_url'`, `'model'`, `'title_status'`, `'vin'`, `'size'`, `'image_url'`, `'description'`, `‘county’`,`'lat'` e `'long'`. O método utilizado para esse fim será o `.drop()`. 


```python
atributos_elimanacao_colunas = ['url', 'region', 'region_url', 'model', 'title_status', 'vin', 'size', 'image_url', 'description','county','lat', 'long']
base_carros = base_carros.drop(columns = atributos_elimanacao_colunas)
base_carros.info()
```

Para uma visualização mais genérica sobre o tamanho do _Dataframe_ que é matematicamente uma matriz de dados poderá ser utilizado o método `.shape`. Ele expressa a quantidade de linhas e colunas do _Dataframe_. Nesse caso aproximaadamente 435 mil linhas e 13 colunas.

```python
base_carros.shape
```

Outra manipulação bastante relevante na fase de limpeza dos dados é em relação aos dados faltantes ( _missing values_ ). São diversas as possibilidades conforme citado no material base. Em um primeiro momento uma opção simplificada é a eliminação daquela linha de registro utilizando o método `.dropna()`.  Veja que a eliminação dos dados do tipo **NaN** ( vazio ou _null_ ) o _Data Frame_ é reduzido para um total 124793 linhas. Isso configura uma redução de cerca 71% da base.

Em algumas ocasiões uma redução da base de dados dessa magnitude poderá afetar o resultado final da Mineração de Dados, logo seria interessante melhorar a base de dados com conhecimento de especialista ou técnicas similares que permitam completar os dados do tipo **NaN**.

```python
base_carros1 = base_carros.dropna()
base_carros1.shape
```

```python
base_carros_dados_faltantes = base_carros1.isnull()
base_carros_dados_faltantes.head()
```

```python
base_carros_dados_faltantes = base_carros1.isnull().sum()
print(base_carros_dados_faltantes)
```
